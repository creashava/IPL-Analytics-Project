```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/ipl_matches.csv")
df.head()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(y=df['match_winner'],
              order=df['match_winner'].value_counts().index)

plt.title("Most Successful IPL Teams")
plt.xlabel("Number of Wins")
plt.ylabel("Teams")
plt.show()
```

```python
toss_result = df['toss_winner'] == df['match_winner']
toss_result.value_counts()
```

```python
toss_result.value_counts().plot.pie(
    autopct='%1.1f%%',
    labels=['Lost After Toss','Won After Toss'],
    figsize=(6,6)
)

plt.title("Impact of Toss on Match Outcome")
plt.ylabel("")
plt.show()
```

```python
toss_result.value_counts().plot.pie(
    autopct='%1.1f%%',
    labels=['Lost After Toss','Won After Toss'],
    figsize=(6,6)
)

plt.title("Impact of Toss on Match Outcome")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x=df['season'])

plt.title("Matches Across Seasons")
plt.xlabel("Season")
plt.ylabel("Matches")
plt.show()
```
